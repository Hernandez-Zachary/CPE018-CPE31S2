*   Reads an image named "nature.jpg"
*   Converts it into grayscale
*   Resizing it to grayscale
*   Saves the result as "processed.jpg




Converts it to grayscale
Resizes it to 300x300 pixels
Saves the result as "processed.jpg"

```python
#import library
import cv2
```

```python
# Step 1: Read the image
img = cv2.imread("nature.jpg")
```

```python
#Saves the result as "processed.jpg
grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#Resizing it to grayscale
resized = cv2.resize(grayscale, (300, 300))
```

```python
#Saves the result as "processed.jpg
cv2.imwrite("processed.jpg", grayscale)
```
